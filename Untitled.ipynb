## Including price data

```python
df = pd.read_csv("/home/smart/Documents/webscraping/realestate.csv")
print(f"{df.columns=}")
df["Bruksareal"][0:10]
```

```python
POSTAL_CODE_PATTERN = re.compile(r", [0-9]{4}")

def extract_postal_code(s) -> int:
    """
    Extracting postalcode from string
    """
    if isinstance(s, str):
        r = POSTAL_CODE_PATTERN.search(s)
        if r is None:
            return None
        try:
            return int(r[0][1:])
        except ValueError as e:
            print(f"{s=}, {e=}")

    return None

def calculate_price_pr_m2(row) -> int:
    """
    Calculate price pr.m**2 from df row
    """
    if ("price" in row) and ("Bruksareal" in row):
        return row["price"] / util.extract_int(row['Bruksareal'][:-1])

    else:
        return None
    
    
```

```python
df["price/m2"] = df["price"]/df["Bruksareal"]
df["postnummer"] = df["location"].apply(extract_postal_code)
```

```python
postal_codes_df = pd.DataFrame()
postal_codes_df = df[["postnummer", "price/m2"]].groupby('postnummer').mean().replace([np.inf, -np.inf], np.nan)
```
