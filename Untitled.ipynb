## Including price data

```python
# Reading Oslo realestate data from my github webscraping repo.
df = pd.read_csv("https://raw.githubusercontent.com/MadBorg/webscraping/master/realestate.csv")
# Prints for a overview
print(f"{df.columns=}")

df["Bruksareal"][0:10]
```

```python
# Regxp pattern for extracting postalcodes from addresses
POSTAL_CODE_PATTERN = re.compile(r", [0-9]{4}")

def extract_postal_code(s: str) -> int:
    """
    Extracting postalcode from string
    """
    if isinstance(s, str):
        r = POSTAL_CODE_PATTERN.search(s)
        if r is None:
            return None
        try:
            return int(r[0][1:])
        except ValueError as e:
            print(f"{s=}, {e=}")

    return None
    
    
```

```python
# Adding relevant columns to the realestate df
df["price/m2"] = df["price"]/df["Bruksareal"]
df["postnummer"] = df["location"].apply(extract_postal_code)
```

```python
# Map reduce to get the mean price for each postal code. Then replacing inf with nan
postal_codes_df = df[["postnummer", "price/m2"]].groupby('postnummer').mean().replace([np.inf, -np.inf], np.nan)
```

```python
# Joinging with the oslo map, to add the prices
price_map = oslo.join(postal_codes_df, on="postnummer")
#print(f"{price_map['price/m2'].min()=}, {price_map['price/m2'].max()=}")
```

```python
# Cleaning
price_map = price_map.replace([np.inf, -np.inf], np.nan)
```

```python

# plot and store fig
f = price_map.plot(column="price/m2", legend=True, missing_kwds={
            "color": "lightgrey",
            "label": "Missing values",
        },
        figsize=(19.20,10.80)
)

# save fig in different formats
f.figure.savefig("static/out/realestate_price_postalcodes.svg")
f.figure.savefig("static/out/realestate_price_postalcodes.jpg")
f.figure.savefig('static/out/realestate_price_postalcodes.eps', format='eps')
```
