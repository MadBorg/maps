```python
import pymongo
import geopandas as gpd
import pandas as pd
import numpy as np
```

## Price per postal code

```python
realests = pymongo.MongoClient("mongodb://localhost:27017").Finn["Realestate2.0"]
```

```python
realests.count_documents({})
```

```python
price_pr_postal_code_pipe = [
    # First selection
    {
        "$project": {
            "price": 1, "postalCode": {
                "$regexFind": {"input": "$address", "regex": r", [0-9]{4}"}
            
            }
        }
    },
    # Fixing postal code
    {"$set": { "postalCode": {"$regexFind": {"input": "$postalCode.match", "regex": r"[0-9]{4}"}}}},
    {"$set": {"postalCode": "$postalCode.match"}},
    # Grouping by postal code and acumilating wiht average.
    {"$group": {
        "_id": "$postalCode",
        "avg_price": {"$avg": "$price"},
    }},
]
```

```python
postal_prices = realests.aggregate(price_pr_postal_code_pipe)
postal_prices_list = list(postal_prices)
postal_prices_df = pd.DataFrame(postal_prices_list) \
    .rename(columns={"_id": "postnummer"}) \
    .dropna()
postal_prices_df["postnummer"] = pd.to_numeric(postal_prices_df["postnummer"], errors='coerce').astype(int)

postal_prices_df.dtypes
```

## Geo

```python
norway_map_gml_path = "static/norway/Basisdata_0000_Norge_3035_Postnummeromrader_GML.gml"

postalcode_map = gpd.read_file(norway_map_gml_path)
postalcode_map.plot(column="postnummer")
```

```python
# Joining postal prices with map
price_map = postalcode_map.merge(postal_prices_df, on="postnummer", how='left')
price_map = price_map.replace([np.inf, -np.inf], np.nan)
```

```python
price_map.columns
```

```python
price_map.plot(column="avg_price", legend=True, missing_kwds={
            "color": "lightgrey",
            "label": "Missing values",
        },
        figsize=(19.20,10.80)
)
```

```python
r = np.array(price_map["avg_price"])
np.sum(r)
```
